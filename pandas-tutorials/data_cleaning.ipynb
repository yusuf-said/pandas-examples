```python
import pandas as pd
```

```python
df = pd.read_excel('/home/yusuf/Masaüstü/coding/pandas-examples/pandas-tutorials/data/Customer_Call_List.xlsx')
```

```python
df.shape
```

```python
df = df.drop_duplicates() #aynı olan satırları temizler.
```

```python
df = df.drop(columns='Not_Useful_Column')
```

```python
df
```

```python
df.shape
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('123/_. ')
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
#apply versiyonu


df['Phone_Number'] = df['Phone_Number'].apply(lambda x:str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--','').replace('Na--','')
```

```python
df
```

```python
df[['street_name','state','zip']] = df['Address'].str.split(',',n=2,expand=True)
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Yes','Y').replace('No','N')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Yes','Y').replace('No','N')
```

```python
df = df.fillna('')
```

```python
df = df.replace('N/a','')
```

```python
for index in df.index:
    if df.loc[index,'Do_Not_Contact'] == 'Y':
        df.drop(index,inplace=True)
```

```python
for index in df.index:
    if df.loc[index,'Phone_Number'] == '':
        df.drop(index,inplace=True)

#df.dropna(subset='Phone_Number',inplace=True)
```
