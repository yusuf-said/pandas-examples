dataset linki: https://www.kaggle.com/datasets/mohammedarfathr/customer-transactions-dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
     filepath_or_buffer='data/unclean_customer_data.csv',
     encoding='utf-8')

print(df.head().to_string())

print(df.shape) #(10200, 6)

print(df.describe()) # datanın medyan vb. değelerini hesaplar.
```

```python
missing_rows = df[df.isnull().any(axis=1)]
missing_rows 
```

```python
df = df.dropna() #eksik değerler fazla olduğundan bu değerleri atıyorum.
```

```python
df.head()
```

```python
missing_rows = df[df.isnull().any(axis=1)]
missing_rows 
```

```python
##veri normalizasyonu buradaki annual_income ve transaciton amount için uygulayacağım

df["Annual_Income"] = (df["Annual_Income"] / df["Annual_Income"].max()).abs()
df["Transaction_Amount"] = (df["Transaction_Amount"] / df["Transaction_Amount"].max()).abs()

```
